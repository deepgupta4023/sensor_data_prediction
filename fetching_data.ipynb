#Next we will insert value for every sensor as a series into our dataframe

```python
df=pd.DataFrame()
index_val=[]
for i in range(5760):  #no of entries in data array to extract timestamps
    index_val.append(records[0]['data'][i]['timestamp'])
df.index=index_val
for variables in range(26): # extracting value for each of 26 variables
    for timestamp in range(5760): # extracting value of each variable for each timestamp
        s=(records[variables]['data'][timestamp]['value'])
    df.insert(variables,column=records[variables]['sensor_name'],value=s)
```
